```python
import zlib
import os

def compress(input_path, output_path, output_filename, reduction_percentage):
    # Eksik boşlukları temizleyin
    output_dir = output_path.strip()

    original_size_mb = os.path.getsize(input_path) / (1024.0 ** 2)

    print(f"Size of the original file: {original_size_mb:.2f} MB")

    with open(input_path, 'rb') as file:
        mob_content = file.read()

    compressed_content = zlib.compress(mob_content)

    # Belirtilen yüzde azaltmaya ulaşana kadar sıkıştırma işlemine devam et
    reduction_factor = float(reduction_percentage) / 100.0
    target_size_mb = original_size_mb - (original_size_mb * reduction_factor)
    while (len(compressed_content) / (1024.0 ** 2)) > target_size_mb and len(compressed_content) < len(mob_content):
        compressed_content = zlib.compress(compressed_content)

    # Dizin var mı kontrol et, yoksa oluştur
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # output_path'i oluştur
    output_path = os.path.join(output_dir, output_filename)

    with open(output_path, 'wb') as output_file:
        output_file.write(compressed_content)

    print(f"Compression complete. Size of the compressed file:: {os.path.getsize(output_path) / (1024.0 ** 2):.2f} MB")

compress(input_path, output_path, output_filename, reduction_percentage)
```
